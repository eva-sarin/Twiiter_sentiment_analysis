```python
tweets_df=tweets_df.drop(['Tweet Id'],axis=1)
tweets_df=tweets_df.drop(['Lat'],axis=1)
tweets_df=tweets_df.drop(['Tweet Location'],axis=1)
tweets_df=tweets_df.drop(['Long'],axis=1)
tweets_df=tweets_df.drop(['Tweet Language'],axis=1)
tweets_df=tweets_df.drop(['User Bio'],axis=1)
tweets_df=tweets_df.drop(['Likes Received'],axis=1)
tweets_df=tweets_df.drop(['Retweets Received'],axis=1)
tweets_df=tweets_df.drop(['User Following'],axis=1)


```
